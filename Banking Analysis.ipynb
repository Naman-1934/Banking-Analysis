```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Banking.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Generate descriptive statistics for the DataFrame

df.describe()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
df.head()
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Examine the distribution of  unique categories in categorical column
categorical_col = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for col in categorical_col:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

# Univariate Analysis

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

# BiVariant Analysis

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
```

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue="Nationality")
```

```python
for col in categorical_col:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(8,4))
    plt.hist(df[col])
    plt.title("Histogram of Occupation Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

# Numerical Analysis

```python
df.head(2)
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Amount of Credit Cards', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending', 'Properties Owned', 'Risk Weighting', 'BRId', 'GenderId', 'IAId']

# Univariate Analysis
plt.figure(figsize=(15,5))
for i,col in enumerate(numerical_cols):
    plt.subplot(5,3,i+1)
    sns.histplot(df[col], kde=True)
plt.show()
```

# Heatmaps

```python
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt='.2f')
plt.title("Heatmap")
plt.show()
```
